```python
import torch
from diffusers import StableDiffusionPipeline

model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

print("Model loaded!")
```

```python
prompt = "A futuristic city with flying cars at sunset, cyberpunk style"

image = pipe(prompt, num_inference_steps=20, guidance_scale=7.5).images[0]
image.save("prodigy_task2_image1.png")
image
```

```python
import gradio as gr
import torch
from diffusers import StableDiffusionPipeline

# तुझा model (already loaded)
def generate_image(prompt):
    image = pipe(prompt, num_inference_steps=25, guidance_scale=7.5).images[0]
    return image

demo = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(label="Text Prompt", placeholder="Futuristic city..."),
    outputs=gr.Image(label="Generated Image"),
    title="🎨 Prodigy Task 2: AI Image Generator",
    description="Stable Diffusion - Text to Image"
)

demo.launch(share=True)
```
